b,

Adjusted R square is another parameter to validate a model. R square is more like correlation, it subracts the remainder of SSE and SST from 1. (R^2 = 1 - sse/sst). 

Sometimes R_square value is misleading, and it also increases if any insignificant feature is added to the model. To overcome this, R_square value was adjusted.
R_sq_adjusted = 1 - [(1-R2)(N-1)]/(N-K-1), where N= no of rows, K = no of cols

R_sq_adjusted wikll alter only if any significant feature is included in the model, and it is more reliable than R_squared metrics.

c,

Variables with P_value < 0.05 are all significant. They are listed below along with their P_value

```python
pval = pd.DataFrame(base_model.pvalues, columns=['P_val'])
pval[pval['P_val']<0.05]
```

d,

We can detect the Multicollinearity with the help of Condition Number from model.summary

If Condn.No < 100, model is very good. No existence of Multi Colllinearity

If Condn. No is between 100 - 1000, then model is good with Moderate collinearity whcih is accepted.

If Condn. No is greater than 1000, then model is bad which has very high multicollinearity.

In our case it is 3.45e+03, and it says that model has very high multi collinearity.

e,

Other important metrics to look for in the model are Durbin Watson test and Prob(Jarque Bera).

Durbin Watson tells us whether there is Auto correlation or NOT. Genrally Durbin Watson ranges from 0 - 4. If it approxiates around 2 for a model, then the model has NO AUTOCORRELATION (which is one of the assumptions of LR).

Jarue_Bera tells us the normality of residues. If the Prob(JB) is less than 0.05 then it suggests that the data isn't normal else Normal.

We have many features in our Base Model and found that all features are not signifiant to build a good model. Hence lets do feature selection using RFE, Recursive Feature Elimination

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
rfe_model = RFE(estimator= linreg) 
# we can also give features_to_select, but since we don't have any uppper limit on the best features let it grow freely.

rfe_model = rfe_model.fit(xtrain, ytrain)
```

```python
feat_index = pd.Series(data= rfe_model.ranking_, index= xtrain.columns)
# it assigns rank for every feature in x

significant_features_rfe = feat_index[feat_index == 1].index
# Assign the features with index == 1, they are the significant freatures

print(significant_features_rfe) # significant_features_rfe
```

```python
# Now we build a model using only the significant features that we got using RFE

xtrain_rfe = xtrain[significant_features_rfe]
model_rfe = sm.OLS(ytrain, sm.add_constant(xtrain_rfe)).fit()
model_rfe.summary()
```

Even after RFE feature selection technique we are getting not sigificant features in our MODEL. Lets do another round of RFE using thoes features that are found to be significant previously.

Also we can see that, the Cond. No has came down by a large extent from 3e+03 to just 36.1. The multicollinearity has been controlled by a greater extent.
